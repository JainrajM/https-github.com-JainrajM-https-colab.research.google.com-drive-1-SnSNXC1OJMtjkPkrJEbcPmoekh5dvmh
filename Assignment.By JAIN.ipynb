```python
import pandas as pd
import matplotlib.pyplot as plt


# var_name=pd.read_csv('path of the file/name_of file') ('/content/iris (1).csv')

dataframe=pd.read_csv("/content/iris (1).csv")
dataframe.head()
```

```python
x=dataframe.SL
y=dataframe.PL
```

```python
plt.scatter(x,y)
plt.xlabel('SL')
plt.ylabel('PL')
plt.title('Scatter plot of SL vs PL')
plt.show
```

```python
plt.hist(dataframe['SL'],bins=10,edgecolor='g',color='r')
plt.xlabel('SL')
plt.ylabel('PL')
plt.title('Histogram ')
plt.show()
```

```python
# Sort the dataset by height in decendung order and select top ten
top_ten=dataframe.sort_values(by='PL', ascending=False).head(10)

# Creat a line plot
plt.figure(figsize=(10,6))
for index, row in top_ten.iterrows():
    plt.plot(row['PW'],row['PL'], marker='o', label=row['SL'])

plt.xlabel('PW')
plt.ylabel('PL')
plt.title('Top ten PL over Years')
plt.legend()
plt.grid(True)
plt.show()
```
